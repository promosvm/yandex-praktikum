Для того что бы постоянно не выполннять одни и теже действия создадим функцию

```python
def lRegression(df, numeric):

    # Масштабирование признаков
    scaler = MinMaxScaler()
    scaler.fit(df[numeric])
    df[numeric] = scaler.transform(df[numeric])

    # Деление на таргет и фичи
    target = df["product"]
    features = df.drop(["product", "id"], axis=1)

    # Деление выборки на тренировочную и тестовую
    features_train, features_test, target_train, target_test = train_test_split(
        features, target, test_size=0.25, random_state=12345
    )

    # Построение модели
    lr = LinearRegression()
    lr.fit(features_train, target_train)
    predictions = lr.predict(features_test)

    # Запись тестовых и предсказанных таргетов в df
    data = {"real_product": target_test, "predicted_product": predictions}
    data_frame = pd.DataFrame(data, columns=["real_product", "predicted_product"])
    # print(data_frame.head(5))

    # Вывод показателей
    rmse = (mean_squared_error(predictions, target_test)) ** (0.5)
    average_product = sum(predictions) / len(predictions)
    print("Average product: {0:.2f}".format(average_product))
    print("RMSE: {0:.2f}".format(rmse))

    # Построение графика
    plt.figure(figsize=(16, 9))
    plt.scatter(data_frame["predicted_product"], data_frame["real_product"])
    plt.title("График зависимости предсказанного значения от фактического")
    plt.xlabel("Предсказанный объем")
    plt.ylabel("Фактический объем")
    plt.plot(
        [target_test.min(), target_test.max()],
        [[target_test.min()], [target_test.max()]],
        color="red",
    )

    return (average_product, rmse, data_frame)
```

Выполним функцию для всех трех регионов и cохраню предсказания и правильные ответы на валидационной
выборке

```python
print("-----Первый регион-------")
average_product_0, rmse_0, df_0_predict = lRegression(df_0, ["f0", "f1", "f2"])
print("-----------------------\n\n")
print("-----Второй регион-------")
average_product_1, rmse_1, df_1_predict = lRegression(df_1, ["f0", "f1", "f2"])
print("-----------------------\n\n")
print("-----Третий регион-------")
average_product_2, rmse_2, df_2_predict = lRegression(df_2, ["f0", "f1", "f2"])
```

## Подготовка к расчету прибыли

**Cправочные данные**

При разведке региона исследуют 500 точек, из которых с помощью
машинного обучения выбирают 200 лучших для разработки.

Бюджет на разработку скважин в регионе — 10 млрд рублей.
При нынешних ценах один баррель сырья приносит 450 рублей дохода.
Доход с каждой единицы продукта составляет 450 тыс. рублей, поскольку
объём указан в тысячах баррелей.

Исходные данные для расчета прибыли

```python
POINTS = 500
COUNT_OF_WELL = 200
BUDGET = 1e7
PRICE_PER_BARREL = 450
```

```python
# цена одной скважины в регионе
price_one_well = BUDGET / COUNT_OF_WELL
print(f"Сумма на разработку одной скважины в регионе {int(price_one_well)}")
```

 Найдем точку безубыточности она равна cумма на разработку одной скважины в регионе деленный на доход с каждой единицы продукта

```python
n = price_one_well / PRICE_PER_BARREL
print("Точка безубыточности: {0:.0f} тыс. баррелей".format(n))
```

Если в скважине объем запаса будет меньше чем 111 тыс. баррелей то скважина будет считатся убыточной

```python
print("Средний запас фактичекой нефти на скважину в регионе 1", df_0["product"].mean())
print("Средний запас фактичекой нефти на скважину в регионе 2", df_1["product"].mean())
print("Средний запас фактичекой нефти на скважину в регионе 3", df_2["product"].mean())
```

Создадим функцию для расчета прибыли на разведанных точках.

```python
def oil_profit_calculation(target, probabilities, count):
    probs_sorted = probabilities.sort_values(ascending=False)
    selected = target[probs_sorted.index][:count]
    return PRICE_PER_BARREL * selected.sum() - BUDGET
```

###   Вывод

Изучены средние фактические запасы по регионам и определена точка безубыточности по объему нефти в скважене. Для первого и третьего региона существует вероятность окупить затраты так как разброс значений вокруг среднего 92.5 и 95 высок, тоесть существует достаточно много скважин, где запасы больше необходимого среднего.

## Расчёт прибыли и рисков

### Функция bootstrap

```python
state = np.random.RandomState(5062020)


def bootstrap_1000(target, probs):
    values = []
    lost = 0
    for i in range(1000):
        target_sample = target.sample(replace=True, random_state=state, n=POINTS)
        probs_sample = probs[target_sample.index]
        profit = oil_profit_calculation(target, probs_sample, COUNT_OF_WELL)
        if profit < 0:
            lost += 1
        values.append(profit)

    values = pd.Series(values)
    lower = values.quantile(0.025)
    upper = values.quantile(0.975)
    average = values.mean()
    risk = lost / 1000
    return average, lower, upper, risk
```

```python
def print_bootstrap(real_product, predicted_product):
    average, lower, upper, risk = bootstrap_1000(real_product, predicted_product)
    print("Данные по региону.")
    print(f"Средняя прибыль по региону = {average:0.2f} тыс.руб.")
    print(f"95% доверительный интервал от {lower:.2f} до {upper:.2f} тыс.руб.")
    print(f"Процент риска {risk:0.2%}")
    return average, lower, upper, risk
```
